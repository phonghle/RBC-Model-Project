```python
import pandas as pd
import fredpy as fp
import numpy as np
import matplotlib.dates as dts
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
# 0. Export path: Set to empty string '' if you want to export data to current directory
export_path = ''


# 1. Import data
data = pd.read_excel(r'C:\Users\phongleh\Desktop\Econ 126  Comp Macro\Project\pwt91.xlsx',sheet_name='Data')
```

```python
data.head()
```

```python
# 2. Create dataset

# Variable equals last year in data
year = data.year.iloc[-1]
```

```python
year
```

```python
# Restrict data to final year
data = data[data['year']==year].reset_index()
```

```python
data.head()
```

```python
# Select columns: 'countrycode','country','cgdpo','emp','hc','ck'
data = data[['year','countrycode','country','cgdpo','emp','hc','ck']]

# Rename columns
data.columns = ['year','country_code','country','gdp','labor','human_capital','physical_capital']

# Drop countries with missing observations
data = data.dropna()
```

```python
# Export data
data[['year','country_code','country','gdp','labor','human_capital','physical_capital']].to_csv(export_path+'cross_country_production.csv',index=False)
```
